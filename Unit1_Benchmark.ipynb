```python
from transformers import pipeline
```

```python
bert_model = "bert-base-uncased"
roberta_model = "roberta-base"
bart_model = "facebook/bart-base"
```

EXPERIMENT-1: TEXT GENERATION

```python
prompt = "The future of Artificial Intelligence is"
```

```python
gen_bert=pipeline("text-generation", model=bert_model)
print("BERT Output:")
print(gen_bert(prompt, max_length=30))
```

```python
gen_roberta=pipeline("text-generation",model=roberta_model)
print("\nRoBERTa Output:")
print(gen_roberta(prompt, max_length=30))
```

```python
gen_bart=pipeline("text-generation", model=bart_model)
print("\nBART Output:")
print(gen_bart(prompt, max_length=30))
```

**BERT**- fails to generate;
**RoBERTa**- fails to generate;
**BART**- performs text generation

BERT and RoBERTa are encoder-only models which do not perform the task of generating next tokens.
BART is an encoder-decoder model trained for generation.



EXPERIMENT-2: MASKED LANGUAGE MODELING

```python
sentence1 = "The goal of Generative AI is to [MASK] new content."
```

```python
fill_bert = pipeline("fill-mask", model=bert_model)
print("BERT:")
print(fill_bert(sentence1))
```

```python
sentence2 = "The goal of Generative AI is to <mask> new content."
```

```python
fill_roberta = pipeline("fill-mask", model=roberta_model)
print("\nRoBERTa:")
print(fill_roberta(sentence2))
```

```python
fill_bart = pipeline("fill-mask", model=bart_model)
print("\nBART:")
print(fill_bart(sentence2))
```

BERT-Predicts the masked word; RoBERTa-predicts the masked word more accurately; BART- does not perform well in predicting the masked word

BERT and RoBERTa are trained mainly on Masked Language Modeling. BART focuses on sentence-level reconstruction and not on single masked tokens.

EXPERIMENT-3: QUESTION ANSWERING

```python
context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
question = "What are the risks?"
```

```python
qa_bert=pipeline("question-answering",model=bert_model)
print("BERT:")
print(qa_bert(question=question, context=context))
```

```python
qa_roberta = pipeline("question-answering", model=roberta_model)

print("\nRoBERTa QA:")
print(qa_roberta(question=question, context=context))
```

```python
qa_bart = pipeline("question-answering", model=bart_model)

print("\nBART QA:")
print(qa_bart(question=question, context=context))
```

All the models produce weak and inconsistent answers.
QA requires task-specific fine-tuning, not just architecture


| Task | Model | Classification (Success/Failure) | Observation (What actually happened?) | Why did this happen? (Architectural Reason) |
|----|----|----|----|----|
| Text Generation | BERT | Failure | Produced gibberish / repeated prompt | BERT is an encoder-only model does not perform the task of generating  next tokens |
| Text Generation | RoBERTa | Failure | Failed similarly to BERT | RoBERTa is also encoder-only and not trained for generation of next tokens |
| Text Generation | BART | Success | Generated coherent continuation | BART is an encoder-decoder model which has a decoder trained for text generation |
| Fill-Mask | BERT | Success | Predicted words like “create”, “generate” | BERT was trained on Masked Language Modeling |
| Fill-Mask | RoBERTa | Success | more accurate predictions of masked words | RoBERTa is also trained on Masked Language Modeling|
| Fill-Mask | BART | Partial / Failure | Less accurate predictions | BART focuses on sentence reconstruction, not token masking |
| QA | BERT | Partial / Failure | Weak or random answer | Model not fine-tuned for QA tasks |
| QA | RoBERTa | Partial / Failure | Inconsistent results | Requires QA fine-tuning  |
| QA | BART | Partial / Failure | Unreliable answer | QA requires task-specific training |
