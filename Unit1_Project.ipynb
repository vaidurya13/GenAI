```python
from transformers import pipeline
```

```python
qa_pipeline = pipeline("question-answering",model="distilbert-base-cased-distilled-squad")
```

```python
context = """
Employees are entitled to 12 casual leaves and 10 sick leaves per calendar year.
Casual leaves must be applied at least one day in advance.
Sick leaves can be taken without prior approval but require a medical certificate
if the leave exceeds two consecutive days. Earned leaves can be carried forward
up to a maximum of 30 days. Maternity leave of 26 weeks is available for eligible
employees as per company policy.
"""
```

```python
questions = [
    "How many sick leaves are allowed?",
    "Is a medical certificate required for sick leave?",
    "How many earned leaves can be carried forward?",
    "How long is maternity leave?"
]
```

```python
for q in questions:
    result = qa_pipeline(question=q, context=context)
    print(f"Q: {q}")
    print(f"A: {result['answer']}\n")
```
